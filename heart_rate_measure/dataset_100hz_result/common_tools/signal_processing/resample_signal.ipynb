```python
import pandas as pd
import os
from scipy.signal import resample

print("Setup Complete")
```

```python
def resample_signal(signal, original_fs, target_fs):
    ratio = target_fs / original_fs
    new_length = int(len(signal) * ratio)
    resampled_signal = resample(signal, new_length)
    return resampled_signal
```

```python
input_folder = 'E:/mimic_dataset/mimic_perform_ppg_signal_csv_125hz'
output_folder = 'E:/mimic_dataset/mimic_perform_ppg_signal_csv_100hz'

os.makedirs(output_folder, exist_ok=True)

for file_name in os.listdir(input_folder):
    if file_name.endswith(".csv"):
        input_path = os.path.join(input_folder, file_name)
        output_path = os.path.join(output_folder, file_name)

        try:
            df = pd.read_csv(input_path, header=None)
            signal = df[0].values

            original_fs = 125  # Hz
            target_fs = 100    # Hz

            resampled_signal = resample_signal(signal, original_fs, target_fs)

            resampled_df = pd.DataFrame(resampled_signal)
            resampled_df.to_csv(output_path, header = False, index = False)

            print(f"Resampled {input_path} and saved at {output_path}")

        except MemoryError as mem_error:
            print(f"Memory error while processing {file_name}: {mem_error}")
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
```
